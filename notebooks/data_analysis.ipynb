# Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt


from data_process import load_data
```

```python
file_path = "../data/accepted_2007_to_2018Q4.csv"
data = load_data(file_path)
```

### Numerical continuous variable: loan_amnt

```python
pd.options.display.float_format = '{:.2f}'.format
loan_amnt_stats = data.loan_amnt.describe()
loan_amnt_stats
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data.loan_amnt, bins=30)
plt.title('Distribution of Loan Amounts')
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.show()
```

### Categorical variable: grade

```python
grades = ["A", "B", "C", "D", "E", "F", "G"]

grade_counts = data['grade'].value_counts()


df_grade_count = pd.DataFrame(grade_counts).sort_values(by='grade').reset_index()
df_grade_count['%'] = df_grade_count['count'] / df_grade_count['count'].sum() * 100

display(df_grade_count)

plt.figure(figsize=(10, 6))
plt.bar(df_grade_count['grade'], df_grade_count['count'])
plt.title('Distribution of Grades')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.show()
```
